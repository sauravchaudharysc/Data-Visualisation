# BarGraph

```python
import matplotlib.pyplot as plt
#to see documentation of it
plt.bar?
```

```python
plt.bar([0,1,2],[10,20,30])
```

```python
plt.bar([0,1,2],[10,20,30])
plt.bar([0,1,2],[20,10,20])
plt.show()
```

```python
#For side by side
import numpy as np
x_coordinates = np.array([0,1,2])
plt.bar(x_coordinates,[10,20,15],width=0.5)
plt.bar(x_coordinates+0.5,[20,10,20],width=0.5)
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Mango","Litchi","Peach"])
plt.bar(x_coordinates+0.5,[20,10,20],width=0.5,label="Present Year")
plt.title("Fruits Price Comparison")
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Mango","Litchi","Peach"])
plt.bar(x_coordinates+0.5,[20,10,20],width=0.5,label="Present Year")
plt.title("Fruits Price Comparison")
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Mango","Litchi","Peach"])
plt.bar(x_coordinates+0.5,[20,10,20],width=0.5,label="Present Year")
plt.title("Fruits Price Comparison")
plt.xlabel("Fruits")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",color="Maroon",tick_label=["Mango","Litchi","Peach"])
plt.bar(x_coordinates+0.5,[20,10,20],width=0.5,label="Present Year",color="Orange")
plt.title("Fruits Price Comparison")
plt.xlabel("Fruits")
plt.ylabel("Price")
#To set y limit
plt.ylim(0,40)
plt.legend()
plt.show()
```
