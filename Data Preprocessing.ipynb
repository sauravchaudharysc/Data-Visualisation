# Normalisation / Standardisation

#### To make the mean of data as zero and standard deviation as 1 . Here we do mean zero so that we can bring the data set to center.


```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dfx = pd.read_csv('X.csv')
dfy = pd.read_csv('Y.csv')
x = dfx.values
y = dfy.values
plt.scatter(x,y)
plt.show()
```

```python
#Normalise
x2=(x-x.mean())/x.std()
plt.scatter(x,y)
plt.scatter(x2,y,color='orange')
plt.show()
```
